#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
data['CATEGORIA_APGAR5'] = ''
```

```python
data.loc[(data['APGAR5'] > 0) & (data['APGAR5'] < 3), 'CATEGORIA_APGAR5'] = 'asfixia grave'
data.loc[(data['APGAR5'] > 3) & (data['APGAR5'] < 5), 'CATEGORIA_APGAR5'] = 'asfixia moderada'
data.loc[(data['APGAR5'] > 5) & (data['APGAR5'] < 7), 'CATEGORIA_APGAR5'] = 'asfixia leve'
data.loc[(data['APGAR5'] > 7) & (data['APGAR5'] < 10),'CATEGORIA_APGAR5'] = 'normal'
data.head()
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
#APGAR5 e SEXO = variável ordenada e descritiva, usar gráfico de barras
ax = sns.countplot(hue='CATEGORIA_APGAR5', x='SEXO', data=data)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha='right')
plt.show()
```

```python
#Porque existe um gráfico sem rótulo?
ax = sns.countplot(x='CATEGORIA_APGAR5', hue='SEXO', data=data)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha='right')
plt.show()
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
data['CATEGORIA_APGAR5'].isna().sum()
```

```python
#tabela de frequência
frq = pd.crosstab(data['GESTACAO'], data['APGAR5'])
frq
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
frq.sum(axis=1)
```

```python
frq_pct = frq.div(frq.sum(axis=1), axis=0)
frq_pct
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
ordem = ['Menos de 22 semanas','22 a 27 semanas','28 a 31 semanas','32 a 36 semanas','37 a 41 semanas','42 semanas e mais']
```

```python
frq_pct = frq_pct.reindex(ordem)
```

```python
frq_pct.plot(kind='bar', figsize=(16, 9))
plt.show()
```

```python
Partos Prematuros ( com menos de 37 semanas)
Os bebês nascidos antes de 37 semanas tendem a ter pontuações APGAR5 mais baixas. Isso é visível pela maior frequência de pontuações nos grupos de menos de 22 semanas até 36 semanas (grupos 1 ao 4).

Partos normais (37 Semanas ou mais)
Os bebês nascidos com 37 semanas ou mais tendem a ter pontuações APGAR5 mais altas. A maioria dos bebês nesta categoria tem pontuações entre 8.0 e 10.0, indicando melhores condições ao nascimento.

Conclusão: os dados apresentados corroboram a afirmação dos médicos de que partos prematuros (menos de 37 semanas) estão associados a menores pontuações APGAR5.
```
